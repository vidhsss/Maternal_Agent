```python
response_data = data.copy()

# Identify rows where 'answer' is not null to count as a response
response_data['response'] = ~response_data['answer'].isnull()

# Group by stage and calculate response rate
response_rate = response_data.groupby('stage')['response'].agg(['sum', 'count'])
response_rate['response_rate'] = (response_rate['sum'] / response_rate['count']) * 100

# Rename columns for clarity
response_rate = response_rate.rename(columns={'sum': 'responses', 'count': 'total_entries'})
```

```python
response_rate 
```

```python
response_data['question'] = response_data['question'].fillna('No Question')  # Replace NaN questions with a placeholder

# Group by stage and question, then calculate response rate
response_rate_with_question = response_data.groupby(['stage', 'question'])['response'].agg(['sum', 'count'])
response_rate_with_question['response_rate'] = (response_rate_with_question['sum'] / response_rate_with_question['count']) * 100

# Rename columns for clarity
response_rate_with_question = response_rate_with_question.rename(
    columns={'sum': 'responses', 'count': 'total_entries'}
)
```

```python
response_rate_with_question
```

```python
sorted_response_weekly= response_rate_with_question.sort_values(
    by='total_entries', ascending=False
)
```
